```python
import numpy as np
from scipy import stats
from scipy.stats import expon
from scipy.stats import chi2
import matplotlib.pyplot as plt
from scipy.stats import norm
from statsmodels.stats import weightstats as stests
import statistics 
from statistics import stdev
path = r"/content/drive/My Drive"
```

```python
# Graficando Chi
df = 3 # parametro de forma.
df2 = 8 # parametro de forma.
chi2 = stats.chi2(df)
t = stats.chi2(df)
t2 = stats.chi2(df2)
x = np.linspace(chi2.ppf(0.01),
                chi2.ppf(0.99), 100)
x = np.arange(0., 30., 0.001)
fp = t.pdf(x) # Función de Densidad
fp2 = t2.pdf(x) # Función de  Densidad
plt.plot(x, fp)
plt.plot(x, fp2)
plt.plot(x, fp,label=r'$\chi^{2}_{3}$',color='black')
plt.plot(x, fp2,label=r'$\chi^{2}_{8}$', color='tab:gray')
plt.title(r'Distribución $\chi^{2}$')
plt.ylabel('Densidad')
plt.xlabel(r'$\chi^{2}$')
plt.legend()
plt.show()
```

```python

chi_obs=6 #valor del chi2
n=10 #grados de libertad

chi2 = stats.chi2(n)
alfa=chi2.sf(chi_obs)

print("La probabilidad acumulada a derecha de un valor chi2 ",chi_obs," es de:",alfa)
```

```python
alfa=0.05 #valor del chi2
n=10 #grados de libertad

chi2 = stats.chi2(n)
chi_obs=chi2.ppf(1-alfa)

print("El valor de chi2 para una probabilidad acumulada a derecha de ",1-alfa," es de:",chi_obs)
```
