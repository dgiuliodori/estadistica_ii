```python
import numpy as np
from scipy import stats
from scipy.stats import expon
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
import scipy.stats as st
path = r"/content/drive/My Drive"
```

```python
mu, sigma = 0, 1 # media y desvio estandar
normal = stats.norm(mu, sigma)

# Graficando t de Student
df = 3 # parametro de forma.
df2 = 8 # parametro de forma.
t = stats.t(df)
t2 = stats.t(df2)
x = np.linspace(t.ppf(0.01),
                t.ppf(0.99), 100)
fp = t.pdf(x) # Funci贸n de Densidad
fp2 = t2.pdf(x) # Funci贸n de  Densidad
fp3 = normal.pdf(x) # Funci贸n de Probabilidad
plt.rcParams["font.size"] = "14"
plt.plot(x, fp)
plt.plot(x, fp2)
plt.plot(x, fp3)
plt.plot(x, fp,label=r'$t_{2}$',color='black')
plt.plot(x, fp2,label=r'$t_{8}$', color='tab:gray')
plt.plot(x, fp3,label=r'$Z$', color='tab:orange')
plt.title('Distribuci贸n t de Student')
plt.ylabel('Densidad')
plt.xlabel('x')
plt.legend()
plt.show()
```
