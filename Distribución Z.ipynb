```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from scipy.stats import norm
import scipy.stats as st
```

```python
# Probabilidad
alfa=0.05

normal = stats.norm(0, 1)
x = np.linspace(normal.ppf(0.0001),
                normal.ppf(0.9999), 100)
fp = normal.pdf(x) # Funci贸n de Probabilidad
plt.plot(x, fp,color='black')
plt.ylabel('Densidad')
plt.xlabel('Z')
plt.rcParams["font.size"] = "10"

z=normal.ppf(alfa)
x2 = np.linspace(normal.ppf(0.0001),
                    normal.ppf(alfa), 100)
fp2 = normal.pdf(x2) # Funci贸n de Probabilidad
plt.fill_between(x2, 0, fp2, color='tab:gray')
plt.show()
print("El valor de z para una probabilidad acumulada a izquierda de ",alfa," es de:",z)
```
