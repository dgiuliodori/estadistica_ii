```python
# Graficando F
dfn = 5 # parametro de forma.
dfd = 12 # parametro de forma.
dfn2 = 13 # parametro de forma.
dfd2 = 25 # parametro de forma.
f = stats.f(dfn,dfd)
f2 = stats.f(dfn2,dfd2)
x = np.linspace(f.ppf(0.01),
                f.ppf(0.99), 100)
x = np.arange(0, 7., 0.001)
fp = f.pdf(x) # Función de Densidad
fp2 = f2.pdf(x) # Función de Densidad
plt.plot(x, fp)
plt.plot(x, fp2)
plt.title('Distribución F')
plt.ylabel('Densidad')
plt.xlabel('F')
plt.plot(x, fp,label=r'$F_{5,12}$', color='black')
plt.plot(x, fp2,label=r'$F_{13,25}$',color='tab:gray')
plt.legend()
plt.show()
```

```python
f_obs=6 #valor del chi2
n=10 #grados de libertad numerador
m=10 #grados de libertad denominador

f = stats.f(n,m)
alfa=f.sf(f_obs)

print("La probabilidad acumulada a derecha de un valor F ",f_obs," es de:",alfa)
```
